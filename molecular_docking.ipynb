```python
!pip install rdkit
from rdkit import Chem
from rdkit.Chem import AllChem

# load ligand
mol = Chem.SDMolSupplier("Structure2D_COMPOUND_CID_91440.sdf")[0]

# add hydrogens
mol = Chem.AddHs(mol)

# generate 3D coordinates
AllChem.EmbedMolecule(mol)

# energy minimization (VERY IMPORTANT)
AllChem.UFFOptimizeMolecule(mol)

# save as SDF (3D optimized)
w = Chem.SDWriter("ligand_3D.sdf")
w.write(mol)
w.close()

print("3D ligand created and minimized")
```

```python
from rdkit.Chem import rdmolfiles

pdb_block = Chem.MolToPDBBlock(mol)

with open("ligand.pdb", "w") as f:
    f.write(pdb_block)

print("Ligand PDB created")
```

```python
import numpy as np

ligand_coords = []

with open("docking_out.pdbqt") as f:
    record = False
    for line in f:
        if line.startswith("MODEL 1"):
            record = True
        elif line.startswith("ENDMDL"):
            break

        if record and (line.startswith("ATOM") or line.startswith("HETATM")):
            x = float(line[30:38])
            y = float(line[38:46])
            z = float(line[46:54])
            ligand_coords.append([x, y, z])

ligand_coords = np.array(ligand_coords)
ligand_center = ligand_coords.mean(axis=0)

print(ligand_center)
```

```python
protein_residues = set()

with open("receptor.pdbqt") as f:
    for line in f:
        if line.startswith("ATOM"):
            x = float(line[30:38])
            y = float(line[38:46])
            z = float(line[46:54])

            dist = ((x - ligand_center[0])**2 +
                    (y - ligand_center[1])**2 +
                    (z - ligand_center[2])**2) ** 0.5

            if dist <= 5.0:
                res_name = line[17:20].strip()
                res_id = line[22:26].strip()
                protein_residues.add((res_name, res_id))

print(sorted(protein_residues))
```
